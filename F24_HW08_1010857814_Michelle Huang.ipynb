### (a)
A Classification Decision Tree is a supervised learning algorithm designed to tackle classification problems, where the objective is to categorize data points into one of several predefined classes. This method operates by splitting the data based on decision rules derived from the input features, which are organized in a tree structure; each internal node represents a decision regarding a feature, while each leaf node signifies a class label. This approach finds numerous real-world applications, including medical diagnosis, where it can classify patients as having a disease or not based on their symptoms and test results. Additionally, it is utilized in fraud detection to identify potentially fraudulent transactions based on transaction details, in customer segmentation for targeted marketing, in email filtering to classify emails as spam or not, and in loan approval processes to assess the likelihood of a loan applicant defaulting.

### (b)
A Classification Decision Tree predicts outcomes by following a sequential path of decisions through its structure. It begins at the root node, which contains the initial decision rule based on a specific feature and a threshold— for instance, "Is age > 30?" Depending on the outcome of this decision, the algorithm traverses the tree, moving to the next relevant node and repeating this process until it reaches a leaf node. This leaf node signifies the predicted class, which is typically determined by identifying the majority class among the data points that have reached that particular node. This methodical approach enables the decision tree to systematically categorize data based on the established rules.

### (c) How does multiple linear regression make predictions?
Linear Regression is used for regression problems (predicting continuous values).

It computes a weighted sum of input features:

y= $\beta_0+\beta_1x1+\beta_2x2+...+\beta_nx+\epsilon$


 
𝑦 is the predicted output, xiare input features, $\beta_i$ are the coefficients (weights), and $\epsilon$is the error term.
The primary distinction between the two lies in their methodologies:
Decision Trees employ a sequence of decisions (if-else logic) to classify data into specific categories, while Linear Regression utilizes a mathematical formula—typically represented as a straight line in simpler scenarios—to forecast a continuous outcome.


When it comes to complexity and interactions, Decision Trees inherently capture interactions between features through the decisions made at each node. For instance, a branch might evaluate conditions such as "Age > 30 and Income > $50,000," illustrating a form of interaction. In contrast, Linear Regression requires that any interactions be explicitly defined and included in the model, such as by adding a term like x1*x2 to represent the interaction between two variables.

### Accuracy

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
In email filtering situations where false positives (genuine emails incorrectly classified as spam) and false negatives (spam emails that go unflagged) are both problematic, accuracy serves as a suitable metric for assessing overall performance. This is because accuracy is particularly relevant when the dataset is balanced, with roughly equal numbers of positive and negative cases, and when the consequences of errors—whether false positives or false negatives—are comparable.
    
### Sensitivity

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$
In the context of disease diagnosis, such as cancer screening, high sensitivity is essential to ensure that the majority of actual cancer cases (positives) are identified, even if this results in an increased number of false positives (FP). The rationale behind this approach is that sensitivity is critical in scenarios where failing to detect a positive case (FN) can lead to severe consequences, including delayed treatment or higher mortality rates.
### Specificity

$$\text{Specificity} = \frac{TN}{TN + FP}$$
In drug testing scenarios, such as anti-doping measures in sports, high specificity is vital for reducing false positives (FP), thereby ensuring that athletes are not unjustly accused of drug use. The rationale for prioritizing specificity lies in the fact that the repercussions of a false positive (FP)—including reputational damage or unwarranted penalties—are often more significant than those associated with a false negative (FN).

### Precision

$$\text{Precision} = \frac{TP}{TP + FP}$$
In the realm of fraud detection for financial transactions, high precision is essential to ensure that flagged transactions are genuinely fraudulent, thereby minimizing the inconvenience caused by false alarms to legitimate customers. The rationale for emphasizing precision is that false positives (FP) can be costly or highly disruptive, as they may lead to unnecessary fraud investigations.

### Key Steps in Using a Decision Tree Classifier

#### What does the `.fit()` method do?

The .fit() method trains the DecisionTreeClassifier by utilizing the input features, denoted as X (independent variables), along with the target variable y (dependent variable). During this training process, the model identifies decision rules that effectively distinguish between different classes based on the provided training data.

#### Why is only the training data used for fitting?

The training dataset is employed to educate the model, whereas the test dataset is kept unseen until the evaluation phase. This approach ensures that the model's predictive performance is assessed without bias, providing a clearer picture of how well it generalizes to new, unseen data.

- Otherwise, the prediction will again be "Paper."

````python
from sklearn.metrics import ConfusionMatrixDisplay

# Confusion matrix for clf
cm_disp_clf = ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred_clf, labels=[0, 1]), 
    display_labels=['Paper', 'Hard']
)
cm_disp_clf.plot()

# Confusion matrix for clf2
cm_disp_clf2 = ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred_clf2, labels=[0, 1]), 
    display_labels=['Paper', 'Hard']
)
cm_disp_clf2.plot()
````

````python
import numpy as np

def calculate_metrics(cm):
    # True Positives, False Positives, True Negatives, False Negatives
    TN, FP, FN, TP = cm.ravel()
    sensitivity = TP / (TP + FN)  # Recall
    specificity = TN / (TN + FP)
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    return sensitivity, specificity, accuracy

# Confusion Matrices
cm_clf = confusion_matrix(y_test, y_pred_clf, labels=[0, 1])
cm_clf2 = confusion_matrix(y_test, y_pred_clf2, labels=[0, 1])

# Metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)
print(f"clf Sensitivity: {np.round(sensitivity_clf, 3)}")
print(f"clf Specificity: {np.round(specificity_clf, 3)}")
print(f"clf Accuracy: {np.round(accuracy_clf, 3)}")

# Metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)
print(f"clf2 Sensitivity: {np.round(sensitivity_clf2, 3)}")
print(f"clf2 Specificity: {np.round(specificity_clf2, 3)}")
print(f"clf2 Accuracy: {np.round(accuracy_clf2, 3)}")
````

Sensitivity (Recall): Measures the proportion of actual Hard books correctly identified.
$$\text{Sensitivity} = \frac{TP}{TP + FN}$$
Specificity: Measures the proportion of Paper books correctly identified.
$$\text{Specificity} = \frac{TN}{TN + FP}$$
Accuracy: Overall correctness of the model's predictions.
$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

Sensitivity and specificity assess the model's ability to differentiate between Hard and Paper categories. Meanwhile, accuracy offers a comprehensive metric of the model's overall performance across all predictions.

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>

But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. In a decision tree, feature importance measures the relative impact of each predictor variable on enhancing the model's predictive accuracy. It is determined by assessing how much each feature decreases impurity (such as Gini impurity or Shannon entropy) at every split in the tree.

````python
import matplotlib.pyplot as plt
import numpy as np

# Extract feature importances and corresponding feature names
feature_importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

# Visualize as a bar plot
plt.figure(figsize=(8, 5))
plt.barh(feature_names, feature_importances, color="skyblue")
plt.xlabel("Feature Importance")
plt.title("Feature Importances for clf2")
plt.show()

# Find the most important feature
most_important_feature = feature_names[np.argmax(feature_importances)]
print(f"The most important predictor variable is: {most_important_feature}")
````

In linear regression, coefficients indicate the precise change in the dependent variable associated with a one-unit increase in the corresponding predictor variable, while keeping all other predictors constant. This direct relationship enhances the interpretability of linear regression, as each coefficient delineates a specific and independent contribution to the outcome.

In contrast, feature importances in decision trees assess the extent to which each predictor variable reduces impurity (such as Gini impurity or entropy) across all splits where it is utilized. Although these importances represent the overall contribution of features, they do not offer the same level of precise interpretability as coefficients. This is because decision trees operate based on hierarchical, conditional rules rather than linear relationships, meaning that the contributions of features are influenced by the tree's structure and the interactions between features.
